```python
import dask.dataframe as dd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
application_train= dd.read_csv("./data/application_train.csv")
a = application_train.shape
print("application_train shape :({},{})".format( a[0].compute(),a[1]))
```

documents fournies / targets

```python
documents_provided=["FLAG_DOCUMENT_{}".format(i) for i in range(2,22)]
```

```python
for i in documents_provided :
    plt.figure(figsize=(30,10))
    sns.countplot(x=i, hue="TARGET", data=application_train.compute())
    plt.show()
```
